```python
import pandas as pd

# Load dataset
df = pd.read_csv(
    '/content/sale/superstore_final_dataset (1).csv',
    encoding='latin1'
)

# Show first 5 rows
print(df.head())
```

```python
# Number of rows and columns
print(df.shape)

# Column names
print(df.columns)

# Dataset information
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
df.drop_duplicates(inplace=True)

print("Duplicates removed")
```

```python
print(df.columns)
```

```python
# Convert Order_Date into datetime
df['Order_Date'] = pd.to_datetime(
    df['Order_Date'],
    dayfirst=True
)

print(df['Order_Date'].head())
```

```python
# Create Month column
df['Month'] = df['Order_Date'].dt.month

# Create Year column
df['Year'] = df['Order_Date'].dt.year

# Check output
print(df[['Order_Date', 'Month', 'Year']].head())
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
# Fill missing Postal_Code values with 0
df['Postal_Code'] = df['Postal_Code'].fillna(0)

print(df.isnull().sum())
```

```python
df.drop_duplicates(inplace=True)

print("Duplicates removed successfully")
```

```python
total_sales = df['Sales'].sum()

print("Total Revenue:", total_sales)
```

```python
average_sales = df['Sales'].mean()

print("Average Sales:", average_sales)
```

```python
max_sale = df['Sales'].max()

print("Maximum Sale:", max_sale)
```

```python
lowest_sale = df['Sales'].min()

print(lowest_sale)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
region_sales = df.groupby('Region')['Sales'].sum()

region_sales.plot(kind='bar')

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.show()
```

Insight:
The West region generated the highest sales revenue compared to other regions.

```python
monthly_sales = df.groupby('Month')['Sales'].sum()

monthly_sales.plot(marker='o')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

```python
category_sales = df.groupby('Category')['Sales'].sum()

category_sales.plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.show()
```

```python
top_products = df.groupby('Product_Name')['Sales'].sum()

top_products = top_products.sort_values(
    ascending=False
).head(10)

print(top_products)
```
